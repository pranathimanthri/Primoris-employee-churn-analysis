```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
allEmp = pd.read_csv('/content/1.AllEmployees.csv')
allEmp['Project Code'] = allEmp['Project Code'].astype('Int64')
allEmp['Last Hire Date'] = pd.to_datetime(allEmp['Last Hire Date'], errors='coerce')
allEmp['Original Hire Date'] = pd.to_datetime(allEmp['Original Hire Date'], errors='coerce')
allEmp['Termination Date'] = pd.to_datetime(allEmp['Termination Date'])
for index, row in allEmp.iterrows():
    if allEmp.iloc[index, 2] == 'T':
          allEmp.iloc[index, 7] = (row['Termination Date'] - row['Last Hire Date']).days
    else:
          allEmp.iloc[index, 7] = (pd.to_datetime('03/18/25') - row['Last Hire Date']).days
allEmp
```

```python
newHires = pd.read_csv('/content/2. NewHires.csv')
newHires['Project Code'] = newHires['Project Code'].astype('Int64')
newHires['Effective Date'] = pd.to_datetime(newHires['Effective Date'])
for index, rows in newHires.iterrows():
  newHires['Days_Since_NewHire'] = pd.to_datetime('03/18/25') - newHires['Effective Date']
newHires['EE #'] = newHires['EE #'].drop_duplicates().astype('Int64')
newHires = newHires.dropna(subset= 'EE #')
newHires = newHires[['EE #', 'FLSA Category']]
newHires
```

```python
promotions = pd.read_csv('/content/3.Promotions.csv')
promotions['Promotion Date'] = promotions['Effective Date']
promotions = promotions[['EE #', 'FLSA Category', 'Promotion Date']]
promotions['Promotion Date'] = pd.to_datetime(promotions['Promotion Date'])
promotions
```

```python
demotions = pd.read_csv('/content/4.Demotions.csv')
demotions['Demotion Date'] = pd.to_datetime(demotions['Effective Date'])
demotions = demotions[['EE #', 'FLSA Category', 'Demotion Date']]
demotions
```

```python
rehires = pd.read_csv('/content/5. Rehires.csv')
rehires['Rehire Date'] = pd.to_datetime(rehires['RehireDate'])
rehires = rehires[['EE #', 'FLSA Category', 'Rehire Date']]
rehires
```

```python
transferIns = pd.read_csv('/content/6.TransferIn.csv')
transferIns['EE #'] = transferIns['ee_#']
transferIns['FLSA Category'] = transferIns['flsa_category']
transferIns['Transfer In Date'] = pd.to_datetime(transferIns['effective_date'])
transferIns = transferIns[['EE #', 'FLSA Category', 'Transfer In Date']]
transferIns
```

```python
transferOuts = pd.read_csv('/content/7. TransferOut.csv')
transferOuts['Transfer Out Date'] = pd.to_datetime(transferOuts['TransferOutDate'])
transferOuts = transferOuts[['EE #', 'FLSA Category', 'Transfer Out Date']]
transferOuts
```

```python
terms = pd.read_csv('/content/8.Terms.csv')
terms['Termination Date'] = pd.to_datetime(terms['Termination Date'])
terms = terms[['EE #', 'Termination Date']]
terms
```

```python
import pandas as pd

# First, extract FLSA and EE # from all secondary DataFrames
flsa_sources = [
    promotions[['EE #', 'FLSA Category']],
    demotions[['EE #', 'FLSA Category']],
    rehires[['EE #', 'FLSA Category']],
    transferIns[['EE #', 'FLSA Category']],
    transferOuts[['EE #', 'FLSA Category']]
]

# Concatenate them all into one DataFrame
flsa_combined = pd.concat(flsa_sources)

# Drop duplicates while keeping the first non-null FLSA per EE #
flsa_combined = flsa_combined.dropna().drop_duplicates(subset='EE #', keep='first')

# Merge FLSA once into allEmp
merged_df = allEmp.merge(flsa_combined, on='EE #', how='left')

# Merge in other data (excluding FLSA columns)
merged_df = merged_df.merge(newHires[['EE #', 'FLSA Category']], on='EE #', how='left')
merged_df = merged_df.merge(promotions[['EE #', 'Promotion Date']], on='EE #', how='left')
merged_df = merged_df.merge(demotions[['EE #', 'Demotion Date']], on='EE #', how='left')
merged_df = merged_df.merge(rehires[['EE #', 'Rehire Date']], on='EE #', how='left')
merged_df = merged_df.merge(transferIns[['EE #', 'Transfer In Date']], on='EE #', how='left')
merged_df = merged_df.merge(transferOuts[['EE #', 'Transfer Out Date']], on='EE #', how='left')
merged_df
```

```python
merged_df.drop_duplicates(subset='EE #', keep='last', inplace=True)
merged_df
```

```python
merged_df['FLSA Category_y'].isna().sum()
```

```python
merged_df['FLSA Category'] = merged_df['FLSA Category_x'].fillna(merged_df['FLSA Category_y'])
merged_df.drop(['FLSA Category_x', 'FLSA Category_y'], axis=1, inplace=True)
merged_df
```

```python
merged_df['Days Since Promotion'] = (pd.to_datetime('03/18/25') - merged_df['Promotion Date']).dt.days.astype('Int64')
merged_df['Days Since Demotion'] = (pd.to_datetime('03/18/25') - merged_df['Demotion Date']).dt.days.astype('Int64')
merged_df['Days Since Rehire'] = (pd.to_datetime('03/18/25') - merged_df['Rehire Date']).dt.days.astype('Int64')
merged_df['Days Since Transfer In'] = (pd.to_datetime('03/18/25') - merged_df['Transfer In Date']).dt.days.astype('Int64')
merged_df['Days Since Transfer Out'] = (pd.to_datetime('03/18/25') - merged_df['Transfer Out Date']).dt.days.astype('Int64')
merged_df['Days Since Termination'] = (pd.to_datetime('03/18/25') - merged_df['Termination Date']).dt.days.astype('Int64')
merged_df
```

```python
for index, row in merged_df.iterrows():
  if merged_df.loc[index,'FLSA Category'] == 'Non-Exempt':
    merged_df.loc[index,'FLSA Category'] = 0
  elif merged_df.loc[index,'FLSA Category'] == 'Exempt':
    merged_df.loc[index,'FLSA Category'] = 1
merged_df['FLSA Exempt'] = merged_df['FLSA Category']
merged_df.drop('FLSA Category', axis=1, inplace=True)
merged_df.head(20)
```
